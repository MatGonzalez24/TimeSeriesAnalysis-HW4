```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def generate_y_phi(phi):
    y = [0]
    for i in range(1,100):
        error = np.random.normal(0,1)
        new_y = y[i-1] * phi + error
        y.append(new_y)
    return y
```

```python
def generate_y_theta(theta):
    y_values = [0]
    e_previous = 0

    for t in range(1, 100):
        e_current = np.random.normal(0, 1)
        y_t = e_current + theta * e_previous
        y_values.append(y_t)

        e_previous = e_current

    return y_values
```

```python
phi_values = [-0.9, -0.7, -0.5, -0.3, 0.3, 0.5, 0.7, 0.9]
```

```python
def phi_hat_calc(y):
    dividend = 0
    divisor = 0
    for i in range(1,100):
        dividend += y[i]*y[i-1]
        divisor += y[i-1]**2
    return dividend/divisor
```

```python
for phi in phi_values:
    y_values = generate_y_phi(phi)

    # Plotting y_values
    plt.figure(figsize=(10, 4))
    plt.plot(range(1, 101), y_values, 'b-')
    plt.xlabel('Time (t)')
    plt.ylabel('y')
    plt.title(f'Plot of y_values for phi = {phi}')
    plt.grid(True)
    plt.show()

    # Plotting autocorrelation function
    plt.figure(figsize=(10, 4))
    plot_acf(y_values, lags=20)
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.title(f'Autocorrelation Function of y_values for phi = {phi}')
    plt.grid(True)
    plt.show()

    phi_hat = phi_hat_calc(y_values)
    print(f'Phi_hat is equal to: {phi_hat}')
```

```python
theta_values = [-0.9, -0.5, -0.3, 0.3, 0.5, 0.9]
```

```python
for theta in theta_values:
    y_values = generate_y_theta(theta)

    # Plotting y_values
    plt.figure(figsize=(10, 4))
    plt.plot(range(1, 101), y_values, 'b-')
    plt.xlabel('Time (t)')
    plt.ylabel('y')
    plt.title(f'Plot of y_values for theta = {theta}')
    plt.grid(True)
    plt.show()

    # Plotting autocorrelation function
    plt.figure(figsize=(10, 4))
    plot_acf(y_values, lags=20)
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.title(f'Autocorrelation Function of y_values for theta = {theta}')
    plt.grid(True)
    plt.show()
```
